```python
#20. Write a code to shuffle a given list randomly without using any built-in shuffle functions.

Answer:-

import random
def shuffle_list(input_list):
    shuffled_list = input_list[:]
    n = len(shuffled_list)
    for i in range(n):
        j= random.randint(0,n-1)
        shuffled_list[i],shuffled_list[j]=shuffled_list[j],shuffled_list[i]
        return shuffled_list
my_list=[1,2,3,4,5,6,7,8]
shuffled=shuffle_list(my_list)
print(shuffled)
Results- random number in list :  5
```

```python
#32. Write a code that takes two dictionaries as input and merges them into a single dictionary. If there are common keys, the values should be added together.

Answer:-

dict1 = {"a":3,"b":2,"c":5}
dict2 = {"c":1,"d":4,"b":6}
merge_dict = dict1.copy()

for key,value in dict1.items():
    if key in dict2:
        merge_dict[key]+=value
    else:
       merge_dict[key]=value
print("merges them into a single dictionary :",merge_dict)
Results- merges them into a single dictionary : {'a': 3, 'b': 4, 'c': 10}
```

```python
#35. Write a code that inverts a dictionary, swapping keys and values. Ensure that the inverted dictionary correctly handles cases where multiple keys have the same value by storing the keys as a list in the inverted dictionary.

Answer:-

original_dict = {"a":54,"c":12,"d":9}
inverted_dict = {}

for key,value in original_dict.items():
    if value in inverted_dict:
        inverted_dict[value].append(key)
    else:
        inverted_dict[value]=[key]
print(inverted_dict)
Results- {54: ['a'], 12: ['c'], 9: ['d']}
```
